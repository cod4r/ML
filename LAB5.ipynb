```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(45)
x = np.random.rand(4000,1) *100
```

```python
print(x)
```

```python
Y = 3.5 * x + np.random.randn(4000, 1) * 20
```

```python
print(Y)
```

```python
model = LinearRegression()
model.fit(x, Y)
```

```python
plt.scatter(x, Y)

# Labels and title
plt.xlabel("x values")
plt.ylabel("Y values")
plt.title("Scatter Plot between x and Y")

# Show plot
plt.show()
```

```python
x = np.random.rand(4000)   # 50 random values between 0 and 1
Y = np.random.rand(4000)

# Calculate correlation
correlation = np.corrcoef(x, Y)[0, 1]

print("Correlation between X and Y:", correlation)
```
